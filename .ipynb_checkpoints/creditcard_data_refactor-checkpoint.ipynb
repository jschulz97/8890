```python
import pandas as pd
```

```python
master_df = pd.DataFrame()
master_dict = dict()
```

```python
df = pd.read_csv('data/creditcarddefault.csv', sep=',', header=0, skiprows=1)
targets = [('X1','Limit'),('X3','Education'),('X4','Marriage'),('X5','Age'),('Y','y')]
for i,row in df.iterrows():
    for key in targets:
        try:
            master_dict[key[1]].append(row[key[0]])
        except:
            master_dict[key[1]] = [row[key[0]]]
```

```python
for cls in master_dict.keys():
    print(cls,len(master_dict[cls]))
```
